
# Exercise 1: Create Arrays
# 1. Create a 1D NumPy array containing the numbers 1 through 20.


```python
# 1. Create a 1D NumPy array containing the numbers 1 through 20.
array_1 = np.arange(1, 20)
print("1D Array from 1 to 19:", array_1)
```

2.Reshape the array into a 4*5 matrix

```python
# 2. Reshape a 1D array of size 20 into a 4x5 matrix.
array_11=np.arange(1, 21)
print(array_11)
reshaped_array = array_11.reshape(4, 5)
print("Reshaped 4x5 Array:\n", reshaped_array)
```

3.Finding of MEAN,MEDIAN,MODE

```python


# 3. Find mean, median, and standard deviation
mean_val = np.mean(array_1)
median_val = np.median(array_1)
std_val = np.std(array_1)

print("\nMean:", mean_val)
print("Median:", median_val)
print("Standard Deviation:", std_val)


```

4.Extract all even Numbers from the array

```python
# 4. Extract all even numbers
even_numbers = array_11[array_11 % 2 == 0]
print("\nEven Numbers:\n", even_numbers)
```

5.Generate a Random 5*5 matrix 
-Transpose of Random Matrix

```python
# 5. Generate a random 5x5 matrix and compute its transpose
random_matrix = np.random.rand(5, 5)
transpose_matrix = random_matrix.T

print("\nRandom 5x5 Matrix:\n", random_matrix)
print("\nTranspose of Random Matrix:\n", transpose_matrix)






```
