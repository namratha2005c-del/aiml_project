
3.This checks whether any column contains empty values.
It counts how many missing values are present in each column.

```python
# 3. Check for missing values and handle them
print("\nMissing Values in Each Column:")
print(df.isnull().sum())

# Fill missing values with 0
df.fillna(0, inplace=True)
```

4. df[["Math", "Science", "English"]].mean(axis=1)

Selects subject columns.

axis=1 means row-wise calculation.

Finds average for each student.

df["Average"] =

Creates new column called "Average

```python
# 4. Calculate average marks of each student
df["Average"] = df.groupby('Student_ID')['Marks'].transform('mean')
print("\nData with Average Marks:")
print(df)
```

5. df["Average"].idxmax()

Finds index of highest average value.

df.loc[]

Retrieves complete row of that student.

```python
# 5. Find the student with highest average marks
top_student = df.loc[df["Average"].idxmax()]
print("\nStudent with Highest Average Marks:")
print(top_student)
```

6. df[df["Math"] > 75]

Filters students.

Condition: Math marks greater than 75.

```python
# 6. Filter students who scored more than 75 in Math
math_above_75 = df[(df["Subject"] == "Math") & (df["Marks"] > 75)]
print("\nStudents who scored more than 75 in Math:")
print(math_above_75)
```

8. np.where(condition, "Pass", "Fail")

If average ≥ 40 → Pass

Else → Fail

Creates new column "Result".

```python
# 7. Add a new column 'Result' (Pass/Fail)
df["Result"] = np.where(df["Average"] >= 40, "Pass", "Fail")

print("\nFinal Data with Result Column:")
print(df)
```
