1. Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

2. Load the dataset CSV file

```python


# Load the CSV file
df = pd.read_csv("student_marks_demo.csv")
```

 3. Calculate Average Marks

```python
# Calculate average marks per student
df["Average"] = df.groupby("Student_ID")["Marks"].transform("mean")
```

 4. Create Result Column (Pass/Fail)

```python
# Add Result column (Pass/Fail based on Average >= 40)
df["Result"] = df["Average"].apply(lambda x: "Pass" if x >= 40 else "Fail")
```

##1. Bar Chart – Student Names vs Average Marks

```python
plt.bar(df["Student_ID"], df["Average"])
```

plt.figure() → Creates a new graph window.

plt.bar() → Creates a bar chart.

plt.title() → Adds title to the chart.

plt.xlabel() → Labels X-axis.

plt.ylabel() → Labels Y-axis.

plt.legend() → Shows legend.

plt.xticks(rotation=45) → Rotates names for better visibility.

plt.show() → Displays the chart.

```python
# 1. Bar Chart: Student Names vs Average Marks
plt.figure()
plt.bar(df["Name"], df["Average"])
plt.title("Student Names vs Average Marks")
plt.xlabel("Student Names")
plt.ylabel("Average Marks")
plt.legend(["Average Marks"])
plt.xticks(rotation=45)
plt.show()
```

##2. Pie Chart – Pass vs Fail Distribution

```python
# 2. Pie Chart: Pass vs Fail Distribution
result_counts = df["Result"].value_counts()

plt.figure()
plt.pie(result_counts, labels=result_counts.index, autopct="%1.1f%%")
plt.title("Pass vs Fail Distribution")
plt.legend(result_counts.index)
plt.show()
```

value_counts() → Counts number of Pass and Fail students.

plt.pie() → Creates a pie chart.

labels= → Shows category names.

autopct="%1.1f%%" → Shows percentage values.

plt.title() → Adds title.

plt.legend() → Displays legend.

plt.show() → Displays the chart

##3. Line Graph – Marks Comparison

```python
# 3. Line Graph: Marks Comparison (Math, Science, English)
df_pivot = df.pivot(index='Name', columns='Subject', values='Marks')

plt.figure()
plt.plot(df_pivot.index, df_pivot["Math"], marker='o', label="Math")
plt.plot(df_pivot.index, df_pivot["Science"], marker='o', label="Science")
plt.plot(df_pivot.index, df_pivot["English"], marker='o', label="English")
plt.title("Marks Comparison (Math, Science, English)")
plt.xlabel("Student Names")
plt.ylabel("Marks")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

plt.plot() → Creates a line graph.

marker='o' → Adds circular points on data.

label= → Names each subject line.

plt.legend() → Displays subject names.

plt.show() → Displays the graph.
